## Investigation Overview


> The goal of this investigation is to look at the relationship between features like employment status, loan term, and Income Range on Borrower Rate


## Dataset Overview

> This data set contains 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others.

> An overview of the data features can be found at (https://docs.google.com/spreadsheets/d/1gDyi_L4UvIrLTEC6Wri5nbaMmkGmLQBk-Yx3z0XDEtI/edit#gid=0)

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
loan  = pd.read_csv("master.csv")
```

## Borrower Rate Distribution

* We can see that the mode of the Borrower Rate at around the 0.31 to 0.32 point
* It stands to show a normal distribution.

```python
#setting the ticks and labels
xticks = np.arange(0, loan.BorrowerRate.max(), 0.05)
labels = ["{:.2f}".format(v) for v in xticks]

#plotting the distribution
plt.figure(figsize = [12, 6])

sb.histplot(data = loan, x = "BorrowerRate", kde = True)
plt.xticks(xticks, labels)
plt.title("Borrower Rate distribution", fontsize = 16)
plt.xlim((0.00, 0.40))
plt.ylabel("Frequency", fontsize = 14)
plt.xlabel("Borrowe Rate", fontsize = 14);
```

## How does Income range affect borrower rate

> It is evident that the highest income groups have the lowest borrower rates, which calls for further investigation

```python
plt.figure(figsize = [20, 8])
sb.boxplot(data = loan, y = "IncomeRange", x = "BorrowerRate", color = sb.color_palette()[2])
plt.title("Income Range by Borrower Rate", fontsize = 16)
plt.xlabel("Borrower Rate", fontsize = 14)
plt.ylabel("Income Range", fontsize = 14);
```

## What is the relationship between employment status, loan term and borrower rate

> The loanees with the shortest loan terms have the lowest borrower rates, with the lowest beign from the fulltime employed loanees. The highest borrower rate is taken by the loanees without employments, with a 36 month plan

```python
plt.figure(figsize = [14, 7])

sb.lineplot(data = loan, x = "EmploymentStatus", y = "BorrowerRate", hue = "Term")
plt.title("")
plt.xlabel("Employment Status", fontsize = 14)
plt.ylabel("Borrower Rate", fontsize = 14)
plt.legend(title = "Loan Term (months)");
```
